```python
import img2data
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('0458.png').astype(np.uint8)
print(img.shape)
plt.imshow(img);
```

```python
images = img2data.img2digits(img)
print(f'len(images): {len(images)}')

for i, n in enumerate(images, start=1):
    plt.subplot(1,4,i)
    plt.imshow(n, cmap='gray')
```

```python
images = img2data.img2digits(img, (28,28), border=4, reshape=False)
print(f'len(images): {len(images)}, shape:{images[0].shape}')

for i, n in enumerate(images, start=1):
    plt.subplot(1,4,i)
    plt.imshow(n, cmap='gray')
```
